## 3. Cluster Analysis

The main purpose of this project is to perform [cluster analysis](https://en.wikipedia.org/wiki/Cluster_analysis#:~:text=Cluster%20analysis%20or%20clustering%20is,in%20other%20groups%20(clusters).) with the [K-Means](https://towardsdatascience.com/understanding-k-means-clustering-in-machine-learning-6a6e67336aa1) algorithm. 

You can perform many [cluster analysis](http://www.stat.columbia.edu/~madigan/W2025/notes/clustering.pdf) using different variables. If you use a maximum of two variables for each cluster analysis, you can identify cluster labels more clearly.

First, the K-Means algorithm expects you to determine the number of clusters (*n_clusters*). You can determine the optimal number of clusters for each cluster analysis in various ways. In this case, you are expected to use the [Elbow Method](https://en.wikipedia.org/wiki/Elbow_method_(clustering).

Finally, different information is obtained in each analysis. Therefore, different labeling should be done as a result of each cluster analysis. 

Labeling example: 

- **Normal Customers**  -- An Average consumer in terms of spending and Annual Income
- **Spender Customers** --  Annual Income is less but spending high, so can also be treated as potential target customer.

### Clustering based on Age and Spending Score

#### *i. Create a new dataset with two variables of your choice*

```python
df1 = df[['age','spending_score']]
hopkins(df1,df1.shape[0])
```

#### *ii. Determine optimal number of clusters*

```python
from yellowbrick.cluster import KElbowVisualizer
kmeans = KMeans()
visu = KElbowVisualizer(kmeans, k = (2,10))
visu.fit(df1)
visu.poof();
```

#### *iii. Apply K Means*

```python
k_means1 =KMeans(n_clusters=4).fit(df1)
labels1 = k_means1.labels_
df1['cluster_id'] = labels1
df1.head()
```

```python
from sklearn.metrics import silhouette_score
silhouette_score(df1, k_means1.labels_)
```

#### *iv. Visualizing and Labeling All the Clusters*

```python
plt.scatter(df1.iloc[:,0], df1.iloc[:,1], c = labels1, s = 50, cmap = "viridis")
plt.scatter(k_means1.cluster_centers_[:, 0], k_means1.cluster_centers_[:, 1], s=200, c='blue',alpha=0.5, label = 'Centroids');
```

```python
#for i in fun:
for i in range(2,20):
    k_means1_i =KMeans(n_clusters=i).fit(df1)
    labels1_i = k_means1_i.labels_
    print(i,'-'*3,silhouette_score(df1, k_means1_i.labels_))
    plt.scatter(df1.iloc[:,0], df1.iloc[:,1], c = k_means1_i.labels_, s = 50, cmap = "viridis")
    plt.scatter(k_means1_i.cluster_centers_[:, 0], k_means1_i.cluster_centers_[:, 1], s=200, c='blue',alpha=0.5, label = 'Centroids')
    plt.show();
```

```python
import plotly.graph_objs as go
import plotly as py
py.offline.init_notebook_mode(connected = True)
df3['cluster_id'] =  k_means3.labels_
trace1 = go.Scatter3d(
    x = df3['age'],
    y = df3['spending_score'],
    z = df3['annual_income'],
    mode ='markers',
    marker = dict(color = df3['cluster_id'], size = 10, line = dict(color= df3['cluster_id'], width= 12), opacity = 0.8)) 
data = [trace1]
layout = go.Layout(title= 'Customers\' Segmentation', scene = dict(xaxis = dict(title  = 'x = age'), 
                                                   yaxis = dict(title  = 'y = Spending Score (1-100)'), 
                                                   zaxis = dict(title  = 'z = annual_income')))
fig = go.Figure(data=data, layout=layout)
py.offline.iplot(fig)
plt.show()
```

### Hierarchical Clustering

#### *i. Determine optimal number of clusters using Dendogram*

```python
#SET2
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings('ignore')
from scipy.cluster.hierarchy import dendrogram, linkage

from scipy.cluster.hierarchy import linkage
hc_complete = linkage(df3, 'complete')
hc_ward = linkage(df3, 'ward')

from scipy.cluster.hierarchy import dendrogram
plt.figure(figsize = (10,3))
plt.title('Dendogram - ward')
plt.xlabel('Observation')
plt.ylabel('Distance')
#plt.yticks(np.linspace(0,700,20))
dendrogram(hc_ward, leaf_font_size=10,no_labels=True)
plt.show();
from scipy.cluster.hierarchy import dendrogram
plt.figure(figsize = (10,3))
plt.title('Dendogram - complete')
plt.xlabel('Observation')
plt.ylabel('Distance')
#plt.yticks(np.linspace(0,700,20))
dendrogram(hc_complete, leaf_font_size=10,no_labels=True)
plt.show();
```

#### ii. *Apply Agglomerative Clustering*

#### iii. *Visualizing and Labeling All the Clusters* 

```python
from sklearn.cluster import AgglomerativeClustering
model = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward')
model.fit_predict(df3)
plt.scatter(df3.iloc[:,0], df3.iloc[:,1], c= model.labels_, cmap = 'viridis')
```

```python
from sklearn.metrics import silhouette_score
silhouette_score(df3, model.labels_)
```
